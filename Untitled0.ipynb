```python
import csv
import pandas
import numpy as np
import sqlite3
from sqlalchemy import create_engine
from pathlib import Path
import re
import matplotlib.pyplot as plt
from wordcloud import WordCloud
from scipy import sparse

import nltk
nltk.download('stopwords')
nltk.download('punkt')
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem.snowball import SnowballStemmer

from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.multiclass import OneVsRestClassifier
from sklearn.linear_model import SGDClassifier
from sklearn import metrics
from sklearn.metrics import f1_score,precision_score,recall_score
from sklearn import svm
from sklearn.linear_model import LogisticRegression

DATA_CSV = Path("Train.csv")
DATA_DB = Path("Train.db")
```

```python
# CREATE A SEPARATE TABLE TO STORE PROCESSED DATA
conn = sqlite3.connect(DATA_DB)
cursor = conn.cursor()
cursor.execute("CREATE TABLE IF NOT EXISTS PROCESSED_DATA(question text NOT NULL, code text, tags text NOT NULL, words_pre integer, words_post integer, is_code integer);")
conn.commit()
conn.close()
```

```python
conn_r = sqlite3.connect(DATA_DB)
_data = pandas.read_sql_query("SELECT Title, Body, Tags FROM DATA", conn_r)

conn_w = sqlite3.connect(DATA_DB)
cursor = conn_w.cursor()

stemmer = SnowballStemmer("english")
stop_words = set(stopwords.words('english'))

total_pre_length = 0
total_post_length = 0
total_code_questions = 0
total_questions = _data.shape[0]

for index, row in _data.iterrows():
  is_code = 0
  title, body, tags = row['Title'], row['Body'], row['Tags']

  if '<code>' in body:
    is_code = 1
    total_code_questions += 1

  current_question_length = len(title) + len(body)
  total_pre_length += current_question_length

  # SEPARATE CODE SNIPPETS FROM BODY
  code = str(re.findall('<code>(.*?)</code>', body, flags=re.DOTALL))

  # REMOVE HTML
  cleaned_code = re.sub("<code>(.*?)</code>", '', body, flags=re.DOTALL)
  cleaned_body = re.sub('<.*?>', '', str(cleaned_code))
    
  question = f"{title} {cleaned_body}"
  question = re.sub('[^A-Za-z]+', ' ', question)

  # TOKENIZE
  words = word_tokenize(str(question.lower()))  
  question = ' '.join(str(stemmer.stem(j)) for j in words if j not in stop_words and (len(j)!=1 or j != 'c'))

  cleaned_question_length = len(question)
  total_post_length += cleaned_question_length

  data = (question, code, tags, current_question_length, cleaned_question_length, is_code)
  cursor.execute("INSERT INTO PROCESSED_DATA VALUES (?,?,?,?,?,?)",data)
  
print(f"[Avg length of question before processing] = {total_pre_length/total_questions}")
print(f"[Avg length of question after processing] = {total_post_length/total_questions}")
print(f"[Percentage of quesitons containing code] = {(total_code_questions/total_questions)*100}")

conn_w.commit()
conn_w.close()
conn_r.close()
```

TAGS ANALYSIS

```python
conn = sqlite3.connect(DATA_DB)
tags = pandas.read_sql_query("SELECT tags FROM PROCESSED_DATA", conn)
conn.close()

vectorizer = CountVectorizer(tokenizer = lambda x: x.split())
tags_v = vectorizer.fit_transform(tags['tags'])
print(f"[Number of unique tags] = {tags_v.shape[0]}")
```

```python
total_tags_count = tags["tags"].apply(lambda x: len(x.split()))
tags['tags_count'] = total_tags_count
tags.head()
```

```python
print(f"[Maximum number of tags per question] = {max(tags['tags_count'])}")
print(f"[Minimum number of tags per question] = {min(tags['tags_count'])}")
print(f"[Average number of tags per question] = { sum(tags['tags_count']) / len(tags['tags_count']) }")
```

```python
tags_names = vectorizer.get_feature_names()
tags_freq = tags_v.sum(axis=0).A1

result = dict(zip(tags_names, tags_freq))

tup = dict(result.items())
#Initializing WordCloud using frequencies of tags.
wordcloud = WordCloud(    background_color='black',
                          width=1600,
                          height=800,
                    ).generate_from_frequencies(tup)

fig = plt.figure(figsize=(30,20))
plt.imshow(wordcloud)
plt.axis('off')
plt.tight_layout(pad=0)
fig.savefig("tag.png")
plt.show()
```

APPLY SUITABLE MODEL

```python
sorted_tags = np.argsort(-tags_freq)

def top_n_tags(n):
    multilabel_yn = tags_v[:,sorted_tags[:n]]
    return multilabel_yn

def questionsCovered(n):
    multilabel_yn = top_n_tags(n)
    NonZeroQuestions = multilabel_yn.sum(axis = 1)
    return np.count_nonzero(NonZeroQuestions), NonZeroQuestions
```

```python
con = sqlite3.connect(DATA_DB)
processed_data = pandas.read_sql_query("SELECT * FROM PROCESSED_DATA", con)
print(f"[Size of Processed Data] = {processed_data.shape[0]}")
con.close()
```

```python
# Removing rows from data with no tags
sumOfRows = questionsCovered(500)[1]
RowIndicesZero = np.where(sumOfRows == 0)[0]
data_new = processed_data.drop(labels = RowIndicesZero, axis = 0)
data_new.reset_index(drop = True, inplace = True)
print(f"[Size of new data] = {data_new.shape[0]}")
```

```python
data_new.head()
```

```python
data_tags = top_n_tags(500)
df = pandas.DataFrame(data_tags.toarray())
TagsDF_new = df.drop(labels = RowIndicesZero, axis = 0)
TagsDF_new.reset_index(drop = True, inplace = True)
print("Size of new data = ",TagsDF_new.shape[0])
```

```python
# SPLIT DATA INTO TEST AND TRAIN ( 80:20 )

totalSize = data_new.shape[0]
trainSize = int(0.20*totalSize)

x_train = data_new.head(trainSize)
x_test = data_new.tail(totalSize - trainSize)

allTags = sparse.csr_matrix(TagsDF_new.values)
y_train = allTags[0:trainSize]
y_test = allTags[trainSize:totalSize]

print("Train data shape ",x_train.shape)
print("Train label shape", y_train.shape)
print("Test data shape ",x_test.shape)
print("Test label shape", y_test.shape)
```

```python
from datetime import datetime

start = datetime.now()
vectorizer = TfidfVectorizer(min_df = 0.0001, ngram_range = (1,3), tokenizer = lambda x: x.split(" "))
TrainData = vectorizer.fit_transform(x_train['question'])
TestData = vectorizer.transform(x_test['question'])
print("Time Taken = ",datetime.now()-start)
```
